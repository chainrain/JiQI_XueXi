```python
import pandas as pd
```

```python
# 泰坦尼克号乘客相关信息
data = pd.read_csv("http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt")
```

```python
data
```

```python
# 选取有用的特征值和目标值(字段)
titanic = data[["pclass", "survived", "age", "sex"]]
```

```python
titanic
```

```python
# 查看缺失值,这个方法有可能不会显示有缺失值
titanic.info()
```

```python
titanic.isnull().any()
```

```python
# 填补age的缺失值,按照age的平均值填补,inplace重写
titanic["age"].fillna(titanic["age"].mean(), inplace=True)
```

```python
# 再次检测是否还有缺失值
titanic.isnull().any()
```

```python
# 筛选特征值和目标值
x = titanic[["pclass", "age", "sex"]]  # pclass:仓位 age:年龄 sex:性别
y = titanic["survived"]  # survived:幸存者
```

```python
x
```

```python
y
```

```python
# 转换成字典
x = x.to_dict(orient="records")
```

```python
x
```

```python
# 划分训练和测试数据集
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
# 字典特征抽取
from sklearn.feature_extraction import DictVectorizer
transfer = DictVectorizer()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
x_train.toarray()
```

```python
# 所有不同的可能性
transfer.get_feature_names()
```

## 决策树

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
# 4、决策树预估器流程
# 1）实例化预估器类
estimator = DecisionTreeClassifier(criterion="entropy")

# 2）调用fit进行训练
estimator.fit(x_train, y_train)

# 3）得出预测结果
y_predict = estimator.predict(x_test)
print("预测结果：\n", y_predict)

# 4）模型评估
# 1）对比真实值和预测值
print("对比真实值和预测值:\n", y_predict == y_test)
# 2）直接计算准确率
accuracy = estimator.score(x_test, y_test)
print("准确率为：\n", accuracy)

# 保存树到本地
export_graphviz(estimator, out_file="titanic67.dot", feature_names=transfer.get_feature_names())
```

#### 决策树API当中，如果没有指定max_depth那么会根据信息熵的条件直到最终结束。
#### 这里我们可以指定树的深度来进行限制树的大小

```python
# 1）实例化预估器类
estimator = DecisionTreeClassifier(criterion="entropy", max_depth=5)

# 2）调用fit进行训练
estimator.fit(x_train, y_train)

# 3）得出预测结果
y_predict = estimator.predict(x_test)
print("预测结果：\n", y_predict)

# 4）模型评估
# 1）对比真实值和预测值
print("对比真实值和预测值:\n", y_predict == y_test)
# 2）直接计算准确率
accuracy = estimator.score(x_test, y_test)
print("准确率为：\n", accuracy)

# 保存树到本地
export_graphviz(estimator, out_file="titanic67.dot", feature_names=transfer.get_feature_names())
```

## 随机森林

```python
from sklearn.ensemble import RandomForestClassifier
# 随机森林预估器
estimator = RandomForestClassifier()
# 准备调参字典,定义超参数的选择列表
param_dict = {"n_estimators": [120,200,300,500,800,1200], "max_depth": [5,8,15,25,30]}
```

```python
from sklearn.model_selection import GridSearchCV
estimator = GridSearchCV(estimator, param_grid=param_dict, cv=3)
# 2）调用fit进行训练
estimator.fit(x_train, y_train)

# 3）得出预测结果
y_predict = estimator.predict(x_test)
print("预测结果：\n", y_predict)

# 4）模型评估
# 1）对比真实值和预测值
print("对比真实值和预测值:\n", y_predict == y_test)
# 2）直接计算准确率
accuracy = estimator.score(x_test, y_test)
print("准确率为：\n", accuracy)

# 查看调优结果
print("最佳参数:\n", estimator.best_params_)
print("最佳结果:\n", estimator.best_score_)
print("最佳估计器:\n", estimator.best_estimator_)
print("交叉验证结果:\n", estimator.cv_results_)
```
