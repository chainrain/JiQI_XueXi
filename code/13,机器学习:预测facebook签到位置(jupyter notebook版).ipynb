```python
import pandas as pd
```

```python
data = pd.read_csv("./FBlocation/train.csv")
```

```python
data
```

```python
# 缩小数据范围
fb_data = data.query("x > 0.5 & x < 1.0 & y > 1.0 & y < 1.5")
```

```python
fb_data.head()
```

```python
# 对时间戳做处理
time = pd.to_datetime(fb_data["time"], unit="s")
```

```python
type(time.values[0])
```

```python
# 转换成时间序列类型的数据
time = pd.DatetimeIndex(time)
```

```python
# 选取有用的时间特征,添加字段到fb_data表里
fb_data["weekday"] = time.weekday
fb_data["hour"] = time.hour
fb_data["day"] = time.day
```

```python
fb_data.head()
```

```python
# 去掉签到较少的地点
place_count = fb_data.groupby("place_id").count()["row_id"]
```

```python
# 查看大于10的地点
place_count[place_count > 10]
```

```python
# 去掉签到较少的地点原数据
new_data = fb_data[fb_data["place_id"].isin(place_count[place_count > 10].index)]
```

```python
new_data.head()
```

```python
# 筛选出特征值和目标值
x = new_data[["x", "y", "accuracy", "weekday", "hour", "day"]]  # x轴位置,y轴位置,accuracy:精确度,星期几,小时,天
y = new_data["place_id"]  # 地点
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

# 数据集划分
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier


# 3、标准化
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
# 思考：x_test应该用fit?fit_transform?transform?
x_test = transfer.transform(x_test)

# 4、KNN预估器流程
# 1）实例化预估器类
estimator = KNeighborsClassifier(n_neighbors=3)

# 2）调用fit进行训练
estimator.fit(x_train, y_train)

# 3）得出预测结果
y_predict = estimator.predict(x_test)
print("预测结果：\n", y_predict)

# 4）模型评估
# 1）对比真实值和预测值
print("对比真实值和预测值:\n", y_predict == y_test)
# 2）直接计算准确率
accuracy = estimator.score(x_test, y_test)
print("准确率为：\n", accuracy)
```
