## 癌症分类预测-良／恶性乳腺癌肿瘤预测

```python
import pandas as pd
```

```python
data = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
```

```python
# 列名
column_name = ['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape',
                   'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
                   'Normal Nucleoli', 'Mitoses', 'Class']
```

```python
# 数据读取
data = pd.read_csv(data, names=column_name)
```

```python
data
```

```python
data.isnull().any()
```

```python
import numpy as np
```

```python
# 1、替换
canser = data.replace(to_replace="?", value=np.nan)
```

```python
canser.isnull().any()
```

```python
canser
```

```python
# 2、删除含有缺失值的样本
canser.dropna(inplace=True)
```

```python
canser
```

```python
# 筛选出特征值和目标值
x = canser.iloc[:, 1:-1]
```

```python
x
```

```python
y = canser["Class"]
```

```python
y
```

```python
# 划分数据集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
transfer = StandardScaler()
# 训练
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
estimator = LogisticRegression()
estimator.fit(x_train, y_train)
```

```python
# 查看模型参数
# 回归系数
estimator.coef_
```

```python
# 偏置
estimator.intercept_
```

```python
# 查看预测结果
y_predict = estimator.predict(x_test)
```

```python
y_predict
```

```python
# 普遍的模型评估方法
# 比对
# 计算准确率
accuracy = estimator.score(x_test, y_test)
```

```python
accuracy
```

```python
# 分类报告
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test, y_predict, labels=[2, 4], target_names=["良性", "恶性"])
```

```python
print(report)
```

```python
# 计算AUC指标
from sklearn.metrics import roc_auc_score
y_true = np.where(y_test > 3, 1, 0)
y_true
```

```python
roc_auc_score(y_true, y_predict)
```
